```python
# Importa a biblioteca 'pandas' como 'pd'
import pandas as pd
# Importa 'urlopen' de 'urllib.request'
from urllib.request import urlopen

# Define a URL da base de dados
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wpbc.data'

'''
Attribute information

1) ID number
2) Outcome (R = recur, N = nonrecur)
3) Time (recurrence time if field 2 = R, disease-free time if field 2 = N)
4-33) Ten real-valued features are computed for each cell nucleus:
    a) radius (mean of distances from center to points on the perimeter)
    b) texture (standard deviation of gray-scale values)
    c) perimeter
    d) area
    e) smoothness (local variation in radius lengths)
    f) compactness (perimeter^2 / area - 1.0)
    g) concavity (severity of concave portions of the contour)
    h) concave points (number of concave portions of the contour)
    i) symmetry 
    j) fractal dimension ("coastline approximation" - 1)
The mean, standard error, and "worst" or largest (mean of the three largest values)
of these features were computed for each image, resulting in 30 features. For instance,
field 4 is Mean Radius, field 14 is Radius SE, field 24 is Worst Radius.
Values for features 4-33 are recoded with four significant digits.
34) Tumor size - diameter of the excised tumor in centimeters
35) Lymph node status - number of positive axillary lymph nodes observed at time of surgery
'''

# Define o nome das colunas da base de dados conforme as informaÃ§Ãµes dos atributos
names = ['id_number', 'outcome', 'time', 'radius_mean', 
         'texture_mean', 'perimeter_mean', 'area_mean', 
         'smoothness_mean', 'compactness_mean', 'concavity_mean',
         'concave_points_mean', 'symmetry_mean', 
         'fractal_dimension_mean', 'radius_se', 'texture_se', 
         'perimeter_se', 'area_se', 'smoothness_se', 
         'compactness_se', 'concavity_se', 'concave_points_se', 
         'symmetry_se', 'fractal_dimension_se', 
         'radius_worst', 'texture_worst', 'perimeter_worst',
         'area_worst', 'smoothness_worst', 
         'compactness_worst', 'concavity_worst', 
         'concave_points_worst', 'symmetry_worst', 
         'fractal_dimension_worst', 'tumor_size', 'lymph_node_status']

# Importa os dados da URL da base de dados e define o nome das colunas conforme o passo anterior
data = pd.read_csv(urlopen(data_url), names = names)

# Exibe os 10 primeiros dados
data.head(10)
```

```python
data.describe()
```

```python
# Mapeia a coluna 'outcome' para valores numÃ©ricos, 'R' de recur serÃ¡ 1, 'N' de nonrecur serÃ¡ 0
data['outcome'] = data['outcome'].map({'R':1.0, 'N':0.0})

# Retira os 4 valores null da coluna 'lymph_node_status'
data = data[data['lymph_node_status'] != '?']

data.head(10)
```

```python
# Altera o Ã­ndice da base de dados para a coluna 'id_number'
data.set_index(['id_number'], inplace = True)

data.head(10)
```

```python
# Checa os tipos de valores da base de dados
data.apply(pd.to_numeric).info()
```

```python
# Converte todo o dataframe para float64
data = data.astype('float64')

data.apply(pd.to_numeric).info()
```

```python
data.describe()
```

```python
# Separa as colunas em dependentes e independentes

from sklearn.model_selection import train_test_split

namesrfc = ['radius_mean', 
         'texture_mean', 'perimeter_mean', 'area_mean', 
         'smoothness_mean', 'compactness_mean', 'concavity_mean',
         'concave_points_mean', 'symmetry_mean', 
         'fractal_dimension_mean', 'radius_se', 'texture_se', 
         'perimeter_se', 'area_se', 'smoothness_se', 
         'compactness_se', 'concavity_se', 'concave_points_se', 
         'symmetry_se', 'fractal_dimension_se', 
         'radius_worst', 'texture_worst', 'perimeter_worst',
         'area_worst', 'smoothness_worst', 
         'compactness_worst', 'concavity_worst', 
         'concave_points_worst', 'symmetry_worst', 
         'fractal_dimension_worst', 'tumor_size', 'lymph_node_status']

X = data[namesrfc]  # CaracterÃ­sticas
y = data['outcome']  # RÃ³tulo

# Separa os dados em conjunto de treinamento e conjunto de teste
# 70% das amostras destinadas para o treinamento e 30% para teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Treinamento do modelo
# Importa 'RandomForestClassifier' de 'usklearn.ensemble'
from sklearn.ensemble import RandomForestClassifier

# Cria um classificador gaussiano
clf = RandomForestClassifier(n_estimators = 100)

# Treina o modelo usando o conjunto de treinamento
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
# Importa 'metrics' de 'sklearn'
from sklearn import metrics
# Modelo de precisÃ£o, que diz quÃ£o frequente o classificador estÃ¡ correto
print("PrecisÃ£o:", metrics.accuracy_score(y_test, y_pred))
```

```python
# OrdenaÃ§Ã£o das caracterÃ­sticas importantes pela pontuaÃ§Ã£o
feature_imp = pd.Series(clf.feature_importances_,index=namesrfc).sort_values(ascending=False)
feature_imp
```

```python
# Importa a biblioteca 'matplotlib.pyplot' como 'plt'
import matplotlib.pyplot as plt
# Importa a biblioteca 'seaborn' como 'sns'
import seaborn as sns

%matplotlib inline

# Cria a barra de plotagem
sns.barplot(x = feature_imp, y = feature_imp.index)

# Adiciona os rÃ³tulos ao grÃ¡fico
plt.xlabel('PontuaÃ§Ã£o da importÃ¢ncia das caracterÃ­sticas - ImportÃ¢ncia relativa')
plt.ylabel('CaracterÃ­sticas')
plt.title("VisualizaÃ§Ã£o das caracterÃ­sticas importantes")
plt.show()
```
